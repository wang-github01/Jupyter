# 梯度下降

```python
x_data = [338.,333.,328.,207.,226.,25.,179.,60.,208.,606.]  # 特征值
y_data = [640.,633.,619.,393.,428.,27.,193.,66.,226.,1591.] # 真实数据
# y_data = b + w * x_data
b = -120
w = -4
lr = 0.0000001  # learning rate 学习率
iteration = 100000 # 重复、迭代次数

# 存储用于打印的初始值
b_history = [b]
w_history = [w]

# iteration
for i in range(iteration):
    
    b_grad = 0.0
    w_grad = 0.0
    for n in range(len(x_data)):
        b_grad = b_grad - 2.0 * (y_data[n] - b - w * x_data[n]) * 1.0  # b 的偏微分
        w_grad = w_grad - 2.0 * (y_data[n] - b - w * x_data[n]) * x_data[n] # w 的偏微分
        
     # 跟新参数
    b = b - lr * b_grad
    w = w - lr * w_grad
```

# 了解线性回归代码

```python
import numpy as np
from sklearn.linear_model import LinearRegression
x_data = np.array([1,2,3,5,7]) # x 轴原始数据数据 
y_data = np.array([4,8,9,10,19]) # y 轴原始数据

#转一下维度，sklearn 框架才能识别
x_data = x_data[:,np.newaxis]
y_data = y_data[:,np.newaxis]

model = LinearRegression() # 建立回归模型
model.fit(x_data,y_data) # 开始训练

# 求解f(x) = wx + b
print("w = ",model.coef_[0],"b = ",model.intercept_)
```
